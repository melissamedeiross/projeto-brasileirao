```python
import pandas as pd
```

```python
#carregamento dos dados
df = pd.read_csv(r"C:\Users\mel_p\Downloads\brasileirao (1).csv")

#visualizar os dados
df

```

```python
#criando novas estatísticas cm base nos dados existentes

df["percentual_vitorias"] = (df["won"] / df["played"])* 100
df["media_gols_marcados"] = df["goals"] / df["played"]
df["media_gols_sofridos"] = df["goals_taken"]/df["played"]

#visualizando os dados
df
```

```python
#filtrando o dataset com os campeôes de cada ano
filtro_campeos = df["place"] == 1

#aplicando filtro ao dataset
df_campeoes = df[filtro_campeos]

df_campeoes

```

```python
#descobrindo o time que mais foi campeão
df_campeoes["team"].mode()
```

```python
#analisando estatísticas e padrôes dos campeões
df_campeoes.groupby("team")["place"].count().sort_values(ascending=False)
```

```python
#analisando estatísticas e padrôes dos campeões
df_campeoes.describe()
```

```python
#Qual foi o melhor campeão dos últimos anos?
indice_maior_campeão = df_campeoes["percentual_vitorias"].idxmax()
df_campeoes.loc[indice_maior_campeão]
```

```python
#Qual foi time teve a defesa mais sólida?
indice_defesa = df_campeoes["media_gols_sofridos"].idxmin()
df_campeoes.loc[indice_defesa]
```
